```python
from datetime import datetime
def get_car_info(car_url):
    info_dict = {}
    import requests
    from bs4 import BeautifulSoup
    try:
        response = requests.get(car_url)
        if not response.status_code == 200:
            return info_dict
        result_page = BeautifulSoup(response.content,'html.parser')
        
       
        table = result_page.find('table', {'class': 'table table-sm mb-4'}).find('tbody')
        for tr in table.find_all("tr"):
            td = tr.find_all('td')

            key = td[0].get_text().strip()
            value = td[1].get_text().strip()
            if key == "שנה" or key == "יד":
                value = int(value)
            elif key == "נפח":
                value = int(''.join(filter(str.isdigit, value)))
                
                
              
                
            elif key == "בעלות קודמת":
                prev = ['פרטית', 'ליסינג']
                if value == prev[0]:
                    value = 'P'  # פרטי
                else:
                    value = 'L'  # אחר
                    
                    
            elif key == "בעלות נוכחית":
                curr = ['פרטית', 'ליסינג']
                if value == curr[0]:
                    value = 'P'  # פרטי
                else:
                    value = 'L' # אחר 
                    
                    
                    
            elif key == "ת. הילוכים":
                gear =['אוטומטי', 'הילוכים']
                if value == gear[0]:
                    value = 'O'  # אוטומטי
                else:
                    value = 'G'  # הילוכים
                    
            info_dict[key] = value         
            
         
    

        model_and_price = result_page.find_all('h2', {'class': 'card-title'})
        info_dict['Price'] = model_and_price[1].get_text().strip().replace('₪', '')
        info_dict['model'] = ' '.join(((model_and_price[0].get_text()).split())[1:])
        
        
        picture = result_page.find_all('div', {'class': 'justify-content-center px-1'})
        info_dict['Pic_num'] = int(len(picture))
        
        
        des = result_page.find_all('div', {'class': 'justify-content-center px-1'})
        
        dates= result_page.find_all('div', {'class': 'px-3'})
        cre_date = dates[0].get_text().split()
        Repub_date = dates[1].get_text().split()
        info_dict['Cre_date'] = cre_date[2]
        info_dict['Repub_date'] = Repub_date[3]
        datetime.strptime(car_dict['Cre_date'], "%Y-%m-%d")
        datetime.strptime(car_dict['Repub_date'], "%Y-%m-%d")
        info_dict["manufacturer"] = "מיצובישי"



        return info_dict
    except:
        return info_dict

    
```

```python
def testings(car_link):
    try:
        response = requests.get(car_link)
        if response.status_code != 200:
            return "str"
        
        result_page = BeautifulSoup(response.content, 'html.parser')
        table = result_page.find('div', {'class': 'order-first order-sm-0'}).find('div')
        
        # Extract text from <p> elements with class 'text-word-break'
        text_list = []
        for p in table.find_all('p', class_='text-word-break'):
            text_list.append(p.get_text(strip=True))
        return text_list[0].replace('\n', '').replace('\r', '')
    except Exception as e:
        return str(e)
```
